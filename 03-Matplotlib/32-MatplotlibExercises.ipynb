Siga as instruções para recriar os gráficos usando estes dados:

## Dados

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```


**Importe matplotlib.pyplot como plt e defina% matplotlib inline se você estiver usando o bloco de anotações jupyter. Que comando você usa se não estiver usando o notebook jupyter**

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

%matplotlib inline
```


## Exercício 1

**Siga estas etapas:**
* **Crie um objeto de figura chamado fig usando plt.figure ()**
* **Use add_axes para adicionar um eixo à tela da figura em [0,0,1,1]. Chame esse novo eixo de eixo.**
* **Plote (x, y) nesses eixos e defina os rótulos e títulos para coincidir com o gráfico abaixo:**

```python
fig = plt.figure()
eixo = fig.add_axes([0, 0, 1, 1])


eixo.xaxis.set_major_locator(ticker.MultipleLocator(20))
eixo.xaxis.set_minor_locator(ticker.MultipleLocator(100))

eixo.yaxis.set_major_locator(ticker.MultipleLocator(50))
eixo.yaxis.set_minor_locator(ticker.MultipleLocator(200))


plt.ylim(0, 200)
plt.xlim(0, 100)

plt.margins(x=0, y=0)
eixo.plot(x, y, 'b')
eixo.set_xlabel('x')
eixo.set_ylabel('y')
eixo.set_title('title')
```


## Exercício 2
**Crie um objeto de figura e coloque dois eixos nele, ax1 e ax2. Localizado em [0,0,1,1] e [0,2,0,5, .2, .2], respectivamente.**

```python

fig = plt.figure()


plt.ylim(0, 200)
plt.xlim(0, 100)

ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])

ax2.xaxis.set_major_locator(ticker.MultipleLocator(0.2))
ax2.yaxis.set_major_locator(ticker.MultipleLocator(.2))
```


**Agora plote (x, y) nos dois eixos. E chame seu objeto de figura para mostrá-lo.**

```python
ax1.plot(x, y, 'b')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.plot(x, y, 'r')
ax2.set_xlabel('x')
ax2.set_ylabel('y')

ax2.xaxis.set_major_locator(ticker.MultipleLocator(20))
ax2.yaxis.set_major_locator(ticker.MultipleLocator(50))

ax2.margins(x=0, y=0)
ax1.margins(x=0, y=0)


fig
```


## Exercício 3

**Crie o gráfico abaixo adicionando dois eixos a um objeto de figura em [0,0,1,1] e [0,2,0,5, .4, .4]**

```python
# Cria figura vazia
fig = plt.figure(dpi=200)

ax1 = fig.add_axes([0, 0, 1, 1]) # eixo principal
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4]) # adiciona eixo
```


**Agora use as matrizes x, ye z para recriar o gráfico abaixo. Observe os limites xlimits e y no gráfico inserido:**

```python

#  Figura maior no eixo 1
ax1.plot(x, z, 'b')
ax1.set_xlabel('X')
ax1.set_ylabel('Z')


# Figura inserida no eixo 2
ax2.plot(x, y, 'b')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('zoom');

ax2.set_xlim([20, 22])
ax2.set_ylim([30, 50])


ax1.margins(x=0, y=0)
fig
```

## Exercício 4

**Use plt.subplots (nrows = 1, ncols = 2) para criar o gráfico abaixo.**

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

**Agora plote (x, y) e (x, z) nos eixos. Brinque com a largura de linha e o estilo**

```python
axes[0].plot(x, y, 'b--')
axes[1].plot(x, z, 'r.-')

axes[0].margins(x=0, y=0)
axes[1].margins(x=0, y=0)

fig
```


**Veja se você pode redimensionar o gráfico adicionando o argumento figsize () em plt.subplots () está copiando e colando o código anterior.**

```python

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 2))

axes[0].plot(x, y, 'b--')
axes[1].plot(x, z, 'r.-')
fig
```
